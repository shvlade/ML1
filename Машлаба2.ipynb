```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

pd.set_option('display.max_colwidth', 5000)
url = 'https://habr.com/ru/articles'
response = requests.get(url)
bs = BeautifulSoup(response.text, 'html.parser')

b = int(input('Сколько статей вывести? '))
count = 0
articles_data = []

for article in bs.find_all('h2', class_="tm-title tm-title_h2"):
    if count >= b:
        break

    title = article.a.span.text
    link = 'https://habr.com' + article.a['href']

    article_response = requests.get(link)
    article_soup = BeautifulSoup(article_response.text, 'html.parser')
    article_content = article_soup.find('div', class_='article-formatted-body article-formatted-body article-formatted-body_version-2')

    if article_content:
        article_text = article_content.get_text(strip=True)
    else:
        article_text = "Текст не найден"

    articles_data.append({
        'Заголовок': title,
        'Ссылка': link,
        'Текст': article_text
    })

    count += 1

df = pd.DataFrame(articles_data)
print(df)
```
