```python
import ipywidgets as widgets
import os

def get_and_store_secret(placeholder, output_secret_path):
  password_field = widgets.Password(
      description=f"Enter Key:",
      placeholder=placeholder,
      disabled=False
  )

  output = widgets.Output()

  def on_button_click(b):
    with output:
      token = password_field.value
      with open(os.path.expanduser(output_secret_path), "w") as f:
        f.write(token)
      password_field.close()
      button.close()

  button = widgets.Button(description="Save Key")
  button.on_click(on_button_click)

  display(password_field, button, output)
```

```python
kaggle_user_path = os.path.expanduser("~/.cache/kaggle-user.txt")
get_and_store_secret("KAGGLE_USERNAME", kaggle_user_path)
```

```python
kaggle_key_path = os.path.expanduser("~/.cache/kaggle.token")
get_and_store_secret("KAGGLE_KEY", kaggle_key_path)
```

```python
with open(kaggle_user_path) as f:
    os.environ["KAGGLE_USERNAME"] = f.read().strip()

with open(kaggle_key_path) as f:
    os.environ["KAGGLE_KEY"] = f.read().strip()
```

```python

import pandas as pd

column_translations = {
    'DATE': 'Дата',
    'TX': 'Максимальная температура',
    'Q_TX': 'Качество_макс_температуры',
    'TN': 'Минимальная температура',
    'Q_TN': 'Качество_мин_температуры',
    'TG': 'Средняя температура',
    'Q_TG': 'Качество_ср_температуры',
    'SS': 'Продолжительность_солнечного_сияния',
    'Q_SS': 'Качество_солнечного_сияния',
    'SD': 'Высота_снежного_покрова',
    'RR': 'Осадки',
    'Q_RR': 'Качество_осадков',
    'QQ': 'Давление',
    'Q_QQ': 'Качество_давления',
    'PP': 'Атмосферное_давление',
    'Q_PP': 'Качество_давления',
    'HU': 'Влажность',
    'Q_HU': 'Качество_влажности',
    'CC': 'Облачность',
    'Q_CC': 'Качество_облачности'
}

df = pd.read_csv('london_weather_data_1979_to_2023.csv')
df_renamed = df.rename(columns=column_translations)


df_renamed['Дата'] = pd.to_datetime(df_renamed['Дата'], format='%Y%m%d')

print(f"\nРазмер данных: {df_renamed.shape}")
print(f"Диапазон дат: с {df_renamed['Дата'].min().strftime('%d.%m.%Y')} по {df_renamed['Дата'].max().strftime('%d.%m.%Y')}")


df_renamed.head(50)
```
